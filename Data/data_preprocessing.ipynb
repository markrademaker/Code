## Create Data Structure

BBC

```python
import pandas as pd


episode_titles = [
    "From Pole to Pole",
    "Mountains",
    "Ice Worlds",
    "Great Plains",
    "Jungles",
    "Seasonal Forests",
    "Fresh Water",
    "Ocean Deep",
    "Shallow Seas",
    "Caves",
    "Deserts"
]

file_paths = [f"/Users/markrademaker/Downloads/Work/Scriptie/Data/BBC_Planet_Earth_Dataset/bbc_{i:02}.mp4" for i in range(1, 12)]
df_media = pd.DataFrame({
    'FilePath': file_paths,
    'Title': episode_titles
})
```

```python
import os
folder_path = "/Users/markrademaker/Downloads/Work/Scriptie/Data/BBC_Planet_Earth_Dataset/BBC_Planet_Earth_Dataset/annotations/shots"
df_shots = pd.DataFrame(columns=['Title', 'StartTime_Shot', 'EndTime_Shot'])

for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    # Extract title from filename
    title = filename[3:-4].replace('_', ' ')  # Removes leading numbers and .txt, replaces _ with space
    
    if os.path.isfile(file_path):
        with open(file_path, 'r') as file:
            for line in file:
                start_time, end_time = line.strip().split('\t')
                df_shots = df_shots.append({
                    'Title': title,
                    'StartTime_Shot': int(start_time),
                    'EndTime_Shot': int(end_time)
                }, ignore_index=True)

df_final = pd.merge(df_shots, df_media, on='Title', how='left')
```

## MovieScenes

```python
import os
import pickle

folder_path = '/Users/markrademaker/Downloads/Work/Scriptie/Data/MovieScene'  # Update this path
data_list = []  # List to store data from each pickle file

for filename in os.listdir(folder_path):
    if filename.endswith('.pkl') or filename.endswith('.pickle'):
        file_path = os.path.join(folder_path, filename)
        try:
            with open(file_path, 'rb') as file:
                data = pickle.load(file)
                data_list.append(data)
        except Exception as e:
            print(f"Failed to load {filename}: {e}")
```
